```python
#grouping the data on the basis of Category(ie either movies or Tv shows) column.

groups=data.groupby('Category')
count_Category=groups.Category.count()
count_Category
```

```python
#Showing a Bar Graph for the count of Movies and Tv Shows.

sns.countplot(data,x='Category')
fig=plt.gcf()
fig.set_size_inches(5,5)
plt.title("Category")
```

```python
#Showing a Barplot for ratings.

sns.countplot(data,x='Rating')
sns.countplot(data,x='Rating').set_xticklabels(sns.countplot(data,x='Rating').get_xticklabels(), rotation=90, fontweight='bold')
fig=plt.gcf()
fig.set_size_inches(8,5)
plt.xlabel('Rating',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.title('RATING')
```

```python
#Showing relationship between Category and Rating.

plt.figure(figsize=(10,5))
sns.countplot(x='Rating',hue='Category',data=data,width=1)
plt.title('Relation between Category and Rating')
plt.show()
```
